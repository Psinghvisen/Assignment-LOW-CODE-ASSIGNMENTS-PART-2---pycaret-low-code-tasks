The Kiva Loans Dataset is based on data from Kiva.org, a micro-lending platform where individuals can lend money to entrepreneurs and small businesses across the world, particularly in developing countries. The dataset contains details about the loans, such as loan amount, loan terms, country, sector, activity type, gender of the borrower, and repayment terms, among other factors.

This is a micro-banking loan dataset where each row represents a borrower with their relevant information. Column ‘en’ captures the loan application text of each borrower, and the column ‘status’ represents whether the borrower defaulted or not (default = 1 or no default = 0)

This dataset is valuable for analyzing various socio-economic factors in micro-lending and is often used for projects involving credit risk analysis, loan default predictions, and anomaly detection.

```python
from pycaret.datasets import get_data
data=get_data('kiva')
```

```python
from pycaret.anomaly import*
s=setup(data,session_id=123)
```

```python
from pycaret.anomaly import AnomalyExperiment
exp=AnomalyExperiment()
```

**Create Model**

This function trains an unsupervised anomaly detection model. All the available models can be accessed using the models function.

```python
iforest=create_model('iforest')
iforest
```

```python
models()
```

**Assign Model**

This function assigns anomaly labels to the training data, given a trained model

```python
iforest_anomalies=assign_model(iforest)
iforest_anomalies
```

**Analyze Model**

We can use the plot_model function to analyzes the performance of a trained model on the test set.

iforest is the Isolation Forest model created earlier using the create_model() function. It's commonly used for anomaly detection and works by isolating outliers in the dataset.

t-SNE (t-distributed Stochastic Neighbor Embedding) is a technique used for reducing the dimensionality of data, particularly useful for visualizing high-dimensional datasets.
t-SNE creates a 2D or 3D map from higher-dimensional data, allowing you to see clusters, patterns, or anomalies in a visual way.
Why Use t-SNE in Anomaly Detection?
Anomalies or outliers will appear as points that are isolated or far from the main clusters of data.
t-SNE helps you visually inspect whether there are clear separations between normal data points and anomalies.

```python
plot_model(iforest,plot='tsne')
```

```python
evaluate_model(iforest)
```

**Prediction**

The predict_model function returns Anomaly and Anomaly_Score label as a new column in the input dataframe.

```python
iforest_pred=predict_model(iforest,data=data)
iforest_pred
```
